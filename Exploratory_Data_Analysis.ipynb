```python
import pandas as pd

health_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/ITB/health_data.csv")
display(health_data.head(10))
print (health_data.info())
```

산점도

```python
#산점도 만들기
sns.scatterplot(data=health_data, x="수축기혈압",y="이완기혈압",hue="음주여부")
```

```python
#막대그래프_음주여부에 따른 체중

import seaborn as sns
sns.countplot(data=health_data,x="체중(5Kg 단위)",hue="음주여부")

from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

plt.figure(figsize=(15,4))

#음주를 하는 사람이 대체로 체중이 더 많이 나간다.
```

```python
#막대그래프
#집단별 평균표 만들기
df_health_data = health_data.groupby('흡연상태' , as_index = False).agg( meantri = ('트리글리세라이드', 'mean'))
df_health_data
```

```python
sns.barplot(data = df_health_data, x='흡연상태', y = 'meantri')
#흡연자가 평균 트리글리세라이드 수치가 높은 경향을 보인다. *정상수치 = 데시리터당 135mg 이하
```

```python
#꺽은선그래프_음주여부별 혈색소

sns.pointplot(data=health_data, x="연령대코드(5세단위)",y="혈색소",hue="음주여부",ci=None)

#글씨 선명
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina") 

plt.figure(figsize=(15,4))

#음주를 하는 사람이 하지 않는 사람보다 혈색소가 높다.
#연령대가 높아질수록 혈색소가 감소한다.
```

```python
#상자그림 만들기
sns.boxplot(data = health_data, x = '요단백', y = 'LDL콜레스테롤')

#x축 요단백 수치가 1에서 6으로 커질수록 y축 LDL 콜레스테롤의 최소값~최대값의 범위는 좁아진다.
#요단백 수치가 1~2 일때는 이상치가 보이지만 그 외에는 보이지 않는다.
#요단백 수치가 5~6 일때는 중앙값이 상자의 아래와 위에 치우쳐 있는 형태의 분포를 보인다.
```
